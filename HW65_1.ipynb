```python
# Violi-Jons

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('titanic.jpg')
plt.imshow(img)
plt.show
```

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_image.shape
```

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
face = face_classifier.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=20, minSize=(100,100))
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,0,255), 4)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.imshow(img_rgb)
plt.show()
```

```python
img = cv2.imread('titanic.jpg')
img_rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rotated)
plt.show()
```

```python
gray_img_rotated = cv2.cvtColor(img_rotated, cv2.COLOR_BGR2GRAY)
gray_img_rotated.shape
```

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

face = face_classifier.detectMultiScale(gray_img_rotated, scaleFactor=1.1, minNeighbors=10, minSize=(140,140))

for (x, y, w, h) in face:
    cv2.rectangle(img_rotated, (x, y), (x+w, y+h), (0,0,255), 4)
```

```python
img_rgb = cv2.cvtColor(img_rotated, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,5))
plt.imshow(img_rgb)
plt.show()
```
